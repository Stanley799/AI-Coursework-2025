```python
# A program that converts the number of days to seconds

print("Welcome to the Days to Seconds Converter...")
print("Type 'stop' when you want to quit...")
print()

'''
Awhile loop that keeps the program running
until the user decides to quit the program.
'''
while True:
    answer = input("Enter number of days (or 'stop' to quit): ")

    #if user wants to stop
    if answer.lower() == 'stop': #this converts to lowercase first.
        print("Bye! Thanks for using the calculator!")
        break

    try:
        days = float(answer)

        # Restricts negative days
        if days < 0:
            print("Please enter a positive number!")
            continue

        #calculation
        hours = days * 24
        minutes = hours * 60
        seconds = minutes * 60

        #display answer
        print(f"{days} days = {seconds:,} seconds")

        # Display the breakdown
        print("Here's how we got that:")
        print(f"  Days: {days}")
        print(f"  Hours: {days} x 24 = {hours:,}")
        print(f"  Minutes: {hours:,} x 60 = {minutes:,}")
        print(f"  Seconds: {minutes:,} x 60 = {seconds:,}")
        print() #skips one line for neatness

    except:
        # If they didn't type a number, tell them nicely
        print("That's not a number! Please try again.")
        print()

print("Program finished!")
```

```python
#Programme that converts radius to volume

import math #helps us to get the exact value of pi

print("Welcome to Sphere Volume Calculator!")
print("Please Enter 'stop' to quit.")

#loop that keeps the program running until the user decides to quit.
while True:
    radius_str = input("\nPlease enter the radius of the sphere: ")

    if radius_str.lower() == "stop": #makes it case-insensitive
        print("Thank you for using the calculator...")
        break

    #converts the input into a float.
    try:
        radius = float(radius_str)

        if radius < 0:# If the radius is a negative
            print("Please enter a positive number...")

            continue #This sends the program back to the start of the loop

        # If the number is valid, calculate volume
        volume = (4/3) * math.pi * (radius ** 3)

        print(f"\nThe volume of a sphere with a radius of {radius} is: {volume}")#f-string enables variables to be placed into the text

    except ValueError:
        print("The number you entered is invalid. Please try again:")

print("\n The Program has been Sucessfully Completed...")
```

```python
#Programe that calculates area and perimeter of a square

#function for calculating area
def calculate_area(side):
    area = side * side
    return area

#Function for calculating the perimeter.
def calculate_perimeter(side):
    perimeter = 4 * side
    return perimeter


print("Welcome to the Square Calculator!")
print("To quit, enter 'stop': ")

#While true loop to make the program running until it is stopped
while True:
    user_input = input("\nEnter the side length of the square: ")

    #if the user wants to quit
    if user_input.lower() == "stop":#.lower() makes it case insensitive
        print("Thanks for using the calculator...")

        break


    # when the user types something that is not a number.
    try:
        # converts the user input into a float
        side = float(user_input)

        # ensures the side is not negative
        if side < 0:
            print("The value cannot be a negative number. Please try again.")
            continue # This sends the program back to the start of the loop.

        #  calls the two functions to do the calculations.
        area_of_square = calculate_area(side)
        perimeter_of_square = calculate_perimeter(side)

        #output
        print(f"The area of the square is: {area_of_square}")
        print(f"The perimeter of the square is: {perimeter_of_square}")


    except ValueError:
        print("Invalid input! Please enter a number for the side length.")

# This message is printed only after the loop has finished.
print("Program has stopped...")
```

```python
#Ptogram that checks the case

#function definition
def check_case(char):
    # The .isupper() method checks if a character is uppercase.
    if char.isupper():
        return f"'{char}' is an uppercase character."
    # The .islower() method checks if a character is lowercase.
    elif char.islower():
        return f"'{char}' is a lowercase character."
    # If it's not a letter at all
    else:
        return f"'{char}' is not a letter."

print("Welcome to the Character Case Checker")
print("\n Enter a single letter to check its case, or type 'stop' to quit.")

# The 'while True' loop ensures the program runs continuously.
while True:
    user_input = input("\nEnter a character: ")

    #if the user wants to stop
    if user_input.lower() == "stop":
        print("Thanks for using the checker...")

        break#ends the program

    #Ensures the user enters one character at a time
    if len(user_input) != 1:
        print("Invalid input! Please enter a single character.")
        continue # sends the program back to the start of the loop

    # If the input is a single character
    result = check_case(user_input)

    # output the results
    print(result)

#once the programme is stopped
print("\n Program ended...")
```

```python
#Aprogram that calculates the average for 5 values

#array
numbers = []

#loop for collecting 5 values
print("Please enter 5 numbers to calculate their average.")
for i in range(5):

    while True:
        # Asks the user for a number and increases the count'i' by 1
        user_input = input(f"Enter number {i + 1}: ")


        try:
            num = float(user_input)#converts the input into float

            # Adds the valid number to our 'numbers' list.
            numbers.append(num)


            break

        except ValueError:#if the value entered is not a number
            print("Invalid input. Please enter a valid number.")

#function for adding the numbers
total_sum = sum(numbers)

#function for getting the average
average = total_sum / len(numbers)

#output
print(f"\nThe numbers you entered are: {numbers}")
print(f"The sum of these numbers is: {total_sum}")
print(f"The average of these numbers is: {average}")
```
